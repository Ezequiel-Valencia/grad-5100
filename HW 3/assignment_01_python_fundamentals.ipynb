### 2.0 Write code that implements the following functionality:
- Prompt the user to input their age in years.
- Convert the years to an integer.
- Output the number of years the person is old.
- Then, output the number of days the person is old by converting years to days (assume 365 days per year).
- Then, output the number of hours the person is old by converting days to hours.
- Then, output the number of seconds the person is old by converting hours to seconds.

```python
age = input("Input your age in years please: ")
```

```python
age = int(age)
age_days = age * 365
age_hours = age_days * 24
age_seconds = age_hours * 60
```

```python
print(f"You are {age} years == {age_days} days == {age_hours} hours == {age_seconds} seconds old.")
```
