### 4.0 Write code that implements the following functionality:
#### Guess a Capital Game:
- Create a dictionary to store country-capital pairs, structured as follows:
    - {'France': 'Paris', 'Finland': 'Helsinki', 'Sweden': 'Stockholm', 'Denmark': 'Copenhagen'}
- Prompt the user 3 times, using a loop, where for each iteration, the user is given a country *randomly* selected from the dictionary, and they must guess the capital.
    - Don't forget to import the random module.
- At each iteration, print whether answer is correct, and what it is.
- After the 3 iterations, Output the number of questions the user got correct.

```python
capitals = {'United States': 'D.C', 'France': 'Paris', 'Finland': 'Helsinki', 'Sweden': 'Stockholm', 'Denmark': 'Copenhagen'}
```

```python
import random
selection_set = list(capitals.keys())

correct = 0
for k in range(3):
    country = random.choice(selection_set)
    selection_set.remove(country)
    guess = input(f"For the country {country}, what is its capital: ")
    if guess == capitals[country]:
        print(f"You are correct, the capital of {country} is {guess}.")
        correct += 1
    else:
        print(f"Sorry the correct answer for {country} is {capitals[country]}, not {guess}.")

print(f"You got {correct} out of 3 correct.")
```

### 5.0 Write code that implements the following functionality:
#### Character Frequency Counter
- Prompt the user for a line of text using the he following text for your prompt:  "Write a line of text: "
- Use a dictionary to store frequency count of the letters.
- Output the frequencies, i.e., for each letter encountered, give the number of times it appears in the text.

```python
text = input("Write a line of text: ")
```

```python
freq = {}
for c in text:
    c = c.lower()
    if c not in freq:
        freq[c] = 1
    else:
        freq[c] += 1
pretty_string = ""
for k in freq.keys():
    pretty_string += f"Character: {k}, Frequency: {freq[k]}\n"
print(f"Here are the frequencies for each character in the text: \n[{text}] \n{pretty_string}")
```

### 8.0 Matplotlib - Write code that implements the following:
- Write a function role_dice(), which simulates a roll of 2 dice.  Use the random library to perform the simulation.
- Use a list comprehension to create a list of 500 calls to role_dice().
- Create a histogram with 11 bins of the list.

```python
import random
import matplotlib.pyplot as plt
# Magic function to state plots should be shown directly below cells
%matplotlib inline
```

```python
def role_dice() -> int:
    return random.choice([1, 2, 3, 4, 5, 6])
dice_rolls = [role_dice() for k in range(500)]
```

```python
plt.hist(x=dice_rolls, bins=11)
```
